```python
#read preprocessed file
import pandas as pd
df=pd.read_csv('F:\sapceX\preprocessed_dataset.csv')
df.isnull().sum()
```

```python
#define x and y(class value is a coulmn in my dataset that is target)
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
#install sklearn to split data into train and test
!pip install scikit-learn
from sklearn.model_selection import train_test_split
```

```python
#define x,y train and x,y test 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=101)
```

```python
#import and intiate & train logistic regrassion
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
#define predictions of x test and train model
predictions=logmodel.predict(x_test)
```

```python
#evalute predict with target
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
#define accuracy in this model
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions,normalize=True)
```

```python
#report classification in logistic regrassion model
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
